```python
import pandas as pd
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
results = pd.read_csv('/content/drive/MyDrive/Modeling Satelities Images Building Damaged/models/test_train_6/results_pre_train_grid.csv')
```

```python
def transform_float(x):
  try:
    return float(x)
  except:
    return np.nan
```

```python
results['val_score'] = results['val_score'].apply(lambda x: transform_float(x))
results['dice_val'] = results['dice_val'].apply(lambda x: transform_float(x))
results['f1_score'] = results['f1_score'].apply(lambda x: transform_float(x))
results['F1_0'] = results['F1_0'].apply(lambda x: transform_float(x))
results['F1_1'] = results['F1_1'].apply(lambda x: transform_float(x))
results['F1_2'] = results['F1_2'].apply(lambda x: transform_float(x))
results['F1_3'] = results['F1_3'].apply(lambda x: transform_float(x))
```

```python
top_trains = results.groupby(["layers","ops","k_sizes","shape_types"]).max().reset_index()
```

```python
top_trains.sort_values(["f1_score"],ascending=False).head(5)
```
