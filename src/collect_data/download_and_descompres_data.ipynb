# Download data

```python
for url, sha1 in tqdm(urls):
  response = requests.get(url)
  if hashlib.sha1(response.content).hexdigest() == sha1:
    nome_arquivo = url.split("/")[-1].split("?")[0] # obter o nome do arquivo a partir do URL
    path_completo = path_data + nome_arquivo # adicionar o nome do arquivo ao caminho do diretório

    with open(path_completo, 'wb') as f:
      f.write(response.content)
    print(f"Arquivo {nome_arquivo} baixado e salvo no diretório {path_data}.")
  else:
    print(f"O hash SHA1 do arquivo {nome_arquivo} está incorreto. O download falhou.")
```
